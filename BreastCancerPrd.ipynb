```python
#Feature Engineering
# split feature& target
x= df.drop("diagnosis",axis=1)
y= df["diagnosis"]
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
# Split the data into test & train
from sklearn.model_selection import train_test_split
x_train, x_test,y_train, y_test= train_test_split(x_scaled,y,test_size=0.2,random_state=42)
```

```python
print( x_train.shape,x_test.shape, y_train.shape, y_test.shape)
```

```python
# import the model
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
```

```python
# Fit The Model
lgr= LogisticRegression()
lgr.fit(x_train,y_train)
y_pred1= lgr.predict(x_test)


print(lgr.score(x_train,y_train))
print(lgr.score(x_test,y_test))
```

```python
from sklearn.ensemble import RandomForestClassifier
RdCls= RandomForestClassifier()
RdCls.fit(x_train,y_train)
y_prd2= RdCls.predict(x_test)
print(RdCls.score(x_train,y_train))
print(RdCls.score(x_test,y_test))
```

```python
from sklearn.tree import DecisionTreeClassifier
DsCls= DecisionTreeClassifier()
DsCls.fit(x_train,y_train)
y_prd3= DsCls.predict(x_test)
print(DsCls.score(x_train,y_train))
print(DsCls.score(x_test,y_test))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
GdCls= GradientBoostingClassifier()
GdCls.fit(x_train,y_train)
y_prd4= GdCls.predict(x_test)
print(GdCls.score(x_train,y_train))
print(GdCls.score(x_test,y_test))
```

```python
from sklearn.neural_network import MLPClassifier
MLCls= MLPClassifier()
MLCls.fit(x_train,y_train)
y_prd4= MLCls.predict(x_test)
print(MLCls.score(x_train,y_train))
print(MLCls.score(x_test,y_test))
```

```python
# Model Evaluation
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score, roc_curve

from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.ensemble import RandomForestClassifier


best_model = RandomForestClassifier().fit(x_train, y_train)
y_pred = best_model.predict(x_test)

print(classification_report(y_test, y_pred))

ConfusionMatrixDisplay.from_estimator(best_model, x_test, y_test)
plt.show()

fpr, tpr, _ = roc_curve(y_test, best_model.predict_proba(x_test)[:,1])
plt.plot(fpr, tpr, label="Random Forest (AUC = %.2f)" % roc_auc_score(y_test, best_model.predict_proba(x_test)[:,1]))
plt.plot([0,1],[0,1],'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import pandas as pd

# Dictionary of models
models = {
    "Logistic Regression": LogisticRegression(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "Gradient Boosting": GradientBoostingClassifier(),
    "SVM": SVC(probability=True),
    "Neural Network": MLPClassifier(max_iter=1000)
}

# Store results
results = []

for name, model in models.items():
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    
    results.append([name, accuracy, precision, recall, f1])

# Create DataFrame for better view
results_df = pd.DataFrame(results, columns=["Model", "Accuracy", "Precision", "Recall", "F1-Score"])
print(results_df)
```

```python
import matplotlib.pyplot as plt

# Plot Accuracy, Precision, Recall, F1
metrics = ["Accuracy", "Precision", "Recall", "F1-Score"]

for i, metric in enumerate(metrics[1:], 1):  # start from Precision for multiple plots
    plt.figure(figsize=(8, 5))
    plt.bar(results_df["Model"], results_df[metric], color="skyblue", edgecolor="black")
    plt.title(f"{metric} Comparison of Models", fontsize=14)
    plt.ylabel(metric)
    plt.xticks(rotation=45)
    plt.ylim(0, 1)
    plt.show()

# Plot accuracy separately
plt.figure(figsize=(8, 5))
plt.bar(results_df["Model"], results_df["Accuracy"], color="lightgreen", edgecolor="black")
plt.title("Accuracy Comparison of Models", fontsize=14)
plt.ylabel("Accuracy")
plt.xticks(rotation=45)
plt.ylim(0, 1)
plt.show()
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

best_model = RandomForestClassifier().fit(x_train, y_train)
y_pred = best_model.predict(x_test)

print(classification_report(y_test, y_pred))

ConfusionMatrixDisplay.from_estimator(best_model, x_test, y_test)
plt.show()

fpr, tpr, _ = roc_curve(y_test, best_model.predict_proba(x_test)[:,1])
plt.plot(fpr, tpr, label="Random Forest (AUC = %.2f)" % roc_auc_score(y_test, best_model.predict_proba(x_test)[:,1]))
plt.plot([0,1],[0,1],'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.show()
```

```python
from sklearn.model_selection import  GridSearchCV

param_grid = {
    "n_estimators": [50, 100, 200],
    "max_depth": [None, 5, 10],
    "min_samples_split": [2, 5, 10]
}

grid = GridSearchCV(RandomForestClassifier(), param_grid, cv=5, scoring="accuracy")
grid.fit(x_train, y_train)

print("Best Params:", grid.best_params_)
print("Best Accuracy:", grid.best_score_)
```
